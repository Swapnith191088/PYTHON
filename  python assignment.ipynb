```python
import pandas as pd
```

```python
df = pd.read_csv("/home/cse-lab/Downloads/iris.csv")
```

```python
df = df.dropna()
```

```python
import matplotlib.pyplot as plt
```

```python
df.head(100)
```

```python
varities = df["variety"].value_counts()
varities = dict(varities)
k= ["Iris-Setosa","Iris-Verginica","Iris-Versicolor"]
v= varities.values()
plt.figure(figsize=(10,10))
plt.pie(x=v, labels = k, explode = [0.2,0.0,0.0],autopct = "%1.1f%%",shadow = True)
plt.title("Iris-Species%")
plt.legend()
plt.show()
```

```python
length = df["sepal.length"]
width = df["sepal.width"]
plt.figure(figsize = (8,8))
#print(width)
plt.scatter(length,width)
#plt.legend()
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```
